```python
import nselib
from nselib import capital_market
```

```python
data = capital_market.index_data("NIFTY 50", from_date="05-02-2012", to_date="05-02-2025")
```

```python
data.head()
```

```python
# Save to CSV file with Date as the first column
data.to_csv("NIFTY50_Historical_Data.csv", index=True, index_label="Date")

print("Data saved successfully as 'NIFTY_Historical_Data.csv'")
```
