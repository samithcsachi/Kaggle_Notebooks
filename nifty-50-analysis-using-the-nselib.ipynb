### Importing the libraries

```python
import nselib

from nselib import capital_market 

import warnings

warnings.filterwarnings('ignore')

```

### Checking the trading calendar 

```python
nselib.trading_holiday_calendar()
```

### All equities traded on particular date

```python
capital_market.bhav_copy_equities('27-01-2025')
```

### Block details 

```python
capital_market.block_deals_data(period = '1M')
```

### Delivery Data

```python
capital_market.deliverable_position_data('SBIN', period ='1M')
```

### Delivery Data and Postion Data

```python
capital_market.price_volume_and_deliverable_position_data('HDFCBANK', period ='1M')
```

### Equity List 

```python
capital_market.equity_list()
```

### Index Data

```python
capital_market.index_data('NIFTY 50', period ='1M')
```

### Price Volume Data

```python
capital_market.price_volume_data('BPCL', period = '1M')
```
