```python
from textblob import TextBlob
from newspaper import Article

url = "https://en.wikipedia.org/wiki/Machine_learning"

article = Article(url)

article.download()
article.parse()
article.nlp()

text = article.summary
print("Article Summary: ")
print(text)
```

```python
blob = TextBlob(text)
print("\n Sentiment Polarity:", blob.sentiment.polarity)
```
