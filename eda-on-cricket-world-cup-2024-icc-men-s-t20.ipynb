    ðŸš€ Creaing New Columns for Further Analysis  ðŸš€

```python
df['1st Batting'] = np.where(
    df['Toss Decision'] == 'Batting',  
    df['Toss Winning'],               
    np.where(                         
        df['Toss Winning'] == df['1st Team'], 
        df['2nd Team'],              
        df['1st Team']  ))             


df['1st Bowling'] = np.where(
        df['Toss Decision'] == 'Fielding',  
        df['Toss Winning'],               
        np.where(                         
            df['Toss Winning'] == df['1st Team'], 
            df['2nd Team'],              
            df['1st Team']))
df = df[df['1st Batting'] != df['1st Bowling']]
```

<a id="dataset-analysis"></a>
<p style="padding:30px;background-color:#2e3b4e;margin:0;color:#e8e8e8;font-family:Arial,sans-serif;font-size:250%;text-align:center;border-radius:10px;overflow:hidden;font-weight:800">
    ðŸš€ Functions for Data Visualization  ðŸš€

```python
def OverallStats(df):

        
    Highest_Score_1stinnings = df.groupby('1st Batting')['First Innings Score'].max().reset_index().sort_values(by='First Innings Score', ascending=False)
    
    
    Highest_Score_2ndinnings = df.groupby('1st Bowling')['Second Innings Score'].max().reset_index().sort_values(by='Second Innings Score', ascending=False)

   
    df['Win_Runs'] = np.where(df['Won by'] == 'Runs', df['Winning Margin'], 0)
    df['Win_Wickets'] = np.where(df['Won by'] == 'Wickets', df['Winning Margin'], 0)

   
    Runs = df[df['Win_Runs'] != 0][['Winners', 'Win_Runs']].sort_values(by='Win_Runs', ascending=False)
    Wickets = df[df['Win_Wickets'] != 0][['Winners', 'Win_Wickets']].sort_values(by='Win_Wickets', ascending=False)

    
    Top_10_Score = df.groupby('Top Scorer')['Highest Score'].max().reset_index().sort_values(by='Highest Score', ascending=False).head(10)
    Top_10_Wickets = df.groupby('Best Bowler')['Best Bowler Figure(Wickets Taken)'].max().reset_index().sort_values(by='Best Bowler Figure(Wickets Taken)', ascending=False).head(10)

    
    Most_POM = df['Player Of The Match'].value_counts().head(10)
    Most_Wins = df['Winners'].value_counts()
    Venue = df['Venue'].value_counts()

    
    return Highest_Score_1stinnings, Highest_Score_2ndinnings, Runs, Wickets, Top_10_Wickets, Top_10_Score, Most_POM, Most_Wins, Venue, df

Highest_Score_1stinnings, Highest_Score_2ndinnings, Runs, Wickets, Top_10_Wickets, Top_10_Score, Most_POM, Most_Wins, Venue, df = OverallStats(df)
```

```python
def plot_scatter(data, x_col,y_col,title, x_label, y_label, hue=None, size=None, style=None, palette='viridis', figsize=(12, 10), rotation=45):

    plt.figure(figsize=figsize)
    ax = sns.scatterplot(data=data, x=x_col, y=y_col, hue=hue, size=size, style=style, palette=palette)
    
    

    for i, row in data.iterrows():
        ax.annotate(f'{row[y_col]:.0f}', 
                    (row[x_col], row[y_col]), 
                    ha='center', va='bottom', 
                    fontsize=10, color='black', 
                    xytext=(0, 5), textcoords='offset points')

    plt.xticks(rotation=rotation, ha='right')
    plt.title(title, fontsize=16)
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
def plot_bar(data, x_col, y_col, title, x_label, y_label, palette='viridis', figsize=(12, 10), rotation=45):

    plt.figure(figsize=figsize)
    ax = sns.barplot(data=data, x=x_col, y=y_col, palette=palette)

    for p in ax.patches:
        ax.annotate(f'{p.get_height():.0f}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    fontsize=12, color='black', 
                    xytext=(0, 5), textcoords='offset points')

    plt.xticks(rotation=rotation, ha='right')  
    plt.title(title, fontsize=16)
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
def plot():

    plot_bar(data=Highest_Score_1stinnings, x_col='1st Batting', y_col='First Innings Score',title= 'Highest First Innings Scores by Team', x_label= 'Team',y_label='Maximum First Innings Score', palette='viridis')
    plot_bar(data=Highest_Score_2ndinnings, x_col='1st Bowling', y_col='Second Innings Score',title= 'Highest Second Innings Scores by Team', x_label= 'Team',y_label='Maximum Second Innings Score', palette='viridis')
    plot_scatter(data=Runs, x_col='Winners', y_col='Win_Runs',title= 'Distribution of Winning Runs', x_label= 'Country',y_label='Winning Runs', palette='viridis')
    plot_scatter(data=Wickets, x_col='Winners', y_col='Win_Wickets',title= 'Distribution of Winning Wickets', x_label= 'Country',y_label='Winning Wickets', palette='viridis')
    plot_bar(data=Top_10_Score, x_col='Top Scorer', y_col='Highest Score',title= 'Highest Score Innings', x_label= 'Player',y_label='Highest Score', palette='deep')
    plot_bar(data=Top_10_Wickets, x_col='Best Bowler', y_col='Best Bowler Figure(Wickets Taken)',title= 'Highest Wickets Innings', x_label= 'Player',y_label='Highest Wickets', palette='deep')
    plot_bar(data=df, x_col=Most_POM.index, y_col=Most_POM.values,title= 'Player of the Matches - Top 10', x_label= 'Player',y_label='Number of Matches', palette='pastel')
    plot_bar(data=df, x_col=Most_Wins.index, y_col=Most_Wins.values,title= 'Most Matches Won by Country', x_label= 'Country',y_label='Number of the Matches Won', palette='dark')
    plot_bar(data=df, x_col=Venue.index, y_col=Venue.values,title= 'No of Matches by Venue', x_label= 'Venue',y_label='Number of the Matches', palette='muted')
```

### Plotting the Charts

```python
plot()
```
